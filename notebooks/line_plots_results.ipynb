```python
import numpy as np

from dds_data.results.smc_results import exp_smc_data
from dds_data.results.dds_results import results_full, results
```

```python
def quants(lst):
  x, y = np.quantile(lst, [0.25,0.75])
  return x,y 
```

```python
def process_quantiles(full_dat):
  results = {}
  for key in full_dat:
    results[key] = {}
    results[key]["smc"] = {}
    print(1, key)

    for method in full_dat[key]:
      if "udmp" not in method:
        results[key][method] = {}
        for nm in full_dat[key][method]:
          data= full_dat[key][method][nm]
          lohi = quants(data)
          m = np.median(data)
          results[key][method][nm] = (m, lohi)
    # smc
    for nm in results[key]["ou"]:
      new_key = f"{key}_smc_{nm}"
      dt = exp_smc_data[new_key]
      m, s = np.median(dt), quants(dt)
      results[key]["smc"][nm] = (m, s)
  return results

results_quants = process_quantiles(results_full)
```

```python

import numpy as np

for key in results:
#   print(results[key], key)
  results[key]["smc"] = {}
#   print(1, key)
  for nm in results[key]["ou"]:
    new_key = f"{key}_smc_{nm}"
    dt = exp_smc_data[new_key]
    m, s = np.mean(dt), np.std(dt)
    results[key]["smc"][nm] = (m, s)
```

```python
import matplotlib.pyplot as plt
import numpy as np
problems = list(results.keys())
print(problems)
models = ['ou', 'pis', 'smc']
print(models)

exp_truth_data = {
    "funnel": 0,
    "lgcp": 512.6,
    "lr_sonar": -108.580864,
    "ion": -111.55908,
    "brownian": 1.0297487,
    "vae": -110.009,
    "nice": 0
} 

exp_vi_data = {
    "funnel":-1.3213242,
    "lgcp": 423.8219,
    "lr_sonar": -122.553085,
    "ion":-115.1121,
    "brownian": -0.3516767,
    "vae": -111.99861,
    "nice": -13.4338255
} 

steps_name = ['3.2', '6.4', '12.8', '25.6']
steps_num = [64, 128, 256, 512]

fontsize = 16
for problem in problems:
  # print(results[problem].keys())
  for model in models:
    vals = results[problem][model]
    means = []
    stds = []
    for step_name, step_num in zip(steps_name, steps_num):
      mean, std = vals[step_name]
      # print(model, mean, std)
      means.append(mean)
      stds.append(std)
    means = np.array(means)
    stds = np.array(stds)
    # print("std", problem, model, stds)
    # print("mean", problem, model, means)
    print("mean minus", problem, model, means - stds)
    print("mean plus", problem, model, means + stds)

    upper = means+stds
    lower = means-stds


    # for i in range(len(upper)):
    #   u, l = upper[i], lower[i]
    #   if l > u:
    #     upper[i] = l
    #     lower[i] = u
    
    plt.plot(steps_num, means, label=model)
    print("upper", problem, model, upper)
    print("lower", problem, model, lower)
    plt.fill_between(steps_num, lower, upper, alpha=.1)



    y_true = [exp_truth_data[problem]] * 4
    y_vi = [exp_vi_data[problem]] * 4

  plt.plot(steps_num, y_true, color='purple', linestyle='--', linewidth=0.9, alpha=0.5, label="ground truth")
  plt.plot(steps_num, y_vi, color='peru', linestyle=':', linewidth=2.5, label="vi")
  plt.title(problem, fontsize=fontsize)
  plt.xlabel('Steps', fontsize=fontsize)
  plt.ylabel('lnZ estimate', fontsize=fontsize)
  plt.legend()
  plt.savefig(f"{problem}_lineplot.pdf", bbox_inches='tight', dpi=300)
  plt.show()
```
